```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
```

## Integral Tak Tentu

```python
sp.integrate(x**2, x)
```

```python
sp.integrate(2*x**2 + 2*x + 1, x)
```

```python
sp.integrate(2*sp.exp(2*x+1), x)
```

```python
sp.integrate(sp.exp(x) + sp.sin(y), x)
```

## Integral Tentu

```python
# sp.integrate(f(x), (integral terhadap x, batas min, batas max)) 
sp.integrate(x**2, (x,1,2))    #dalam bilangan pecahan
```

```python
def f(x):
    return x**2
quad(f, 1, 2) [0]              #dalam bilangan decimal
```

```python
#plot bidang integral
def f(x):
    return x**2
x = np.linspace(-2, 5, 1000)
plt.plot(x, f(x))
plt.fill_between(x, f(x), where=[(x>1) and (x<2) for x in x], color='red')
```

```python
x = sp.Symbol('x')
sp.integrate(2*x**2 + x + 1, (x,2,5)) 
```

```python
def f(x):
    return 2*x**2+x+1
x = np.linspace(-2, 5, 1000)
plt.plot(x, f(x))
plt.fill_between(x, f(x), where=[(x>2) and (x<5) for x in x], color='yellow')
```

```python
x = sp.Symbol('x')
sp.integrate(2*sp.exp(2*x), (x,0,2*sp.pi))   #sp.pi masih dlm simbol, np.pi sudah dlm decimal
```

```python
def f(x):
    return 2*np.exp(2*x)
x = np.linspace(1, 5, 1000)
plt.plot(x, f(x))
plt.fill_between(x, f(x), where=[(x>0) and (x<2*sp.pi) for x in x], color='pink')
```

## Plot Sebuah Fungsi

```python
def f(x):
    return np.sin(x)
x = np.linspace(-3, 3, 1000)
plt.plot(x, f(x))
plt.title('my graph')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.axhline(color='black')
plt.axvline(color='black')
```
